Numpy Array V/S List:-

1. In List data of multiple data type can be stored at a time, While in array 
   it only data type need to be selected to perform mathematical calculation.
2. For List no module need to be installed while for numpy module need to be      insatlled.
3. USing List heavy numerical problem(MATRIX OPERATON) cannot be perform,          while numpy is able to perform.
4. More modification option are avialable in list, it is not in array.
5. Numpy Array consume less memory in compare to list.
6. Numpy array is much faster than python list.

```python
%%timeit
x=np.array([1,2,3,6])
for i in x:
    print(i)
```

```python
%%timeit
list =[1,2,3,6]
for i in list:
    print(i)
#[for i in list]
```

# Creating Numpy array using function :  
1. zero's array.
2. one's array.
3. Empty array.
4. array with range of element.
5. Diagonal element filled with 1's.
6. Create array with values that are spaced linearly

ZEROS:

```python
zero_array=np.zeros(3)
print(zero_array)
print ()
zero_array1=np.zeros((3, 4))
print(zero_array1)
```

Ones:

```python
ones_array=np.ones(3)
print(ones_array)
print ()
ones_array1=np.ones((3, 4))
print(ones_array1)
```

Empty:

```python
empty_array=np.empty(3)
print(empty_array)

# The output of this function will depend on the previous array created,
# It will take values from previously created array
# If there is no previous array default it will take zero's value
```

Range:

```python
range_array=np.arange(5)
print(range_array)
```

Diagonal:

```python
diagonal_array=np.eye(3)
print(diagonal_array)
```

```python
diagonal_array=np.eye(3,5)
print(diagonal_array)
```

Linespace:

```python
line_array=np.linspace(0,20,num=5)
print(line_array)
```

# Creating numpy array with random numbers:

1. rand() : It will create a 1-d array with number between 0 & 1
2. randn(): It will crate create array of value close to zero, values may be 
            +ve or -ve.
3. randf() : It fills random number between range [0.0, 1.0).
4. randint():It will genrate values in array between given range.

```python
# 1. 
ran=np.random.rand(5)
print(ran)
```

```python
# 2. 
ran_n=np.random.randn(5)
print(ran_n)
```

```python
# 3.
ran_f=np.random.ranf(4)
print(ran_f)
```

```python
# 4. 
# ran_int=np.random.randint(min, max, number of elements)

rand_int=np.random.randint(1,6,4)
print(rand_int)
```
