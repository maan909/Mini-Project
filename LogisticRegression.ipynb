```python
import pandas as pd
import numpy as np 
from sklearn.model_selection import train_test_split
import matplotlib as plt 
from sklearn.linear_model import LogisticRegression 

from sklearn.metrics import accuracy_score
```

```python
train = pd.read_csv("F:\Mini-Project\Mini-Project\Traffic.csv")
train                   
                
```

```python
train, test = train_test_split(train, test_size=0.2, random_state=42)
print("Shape of train: ", train.shape)
print("Shape of test", test.shape)
```

```python
test
```

```python
train.info()
```

```python
from sklearn.preprocessing import LabelEncoder 
le = LabelEncoder()



train_temp = train
train_temp['Day of the week'] = train_temp['Day of the week'].replace({'Monday':1,'Tuesday':2,'Wednesday':3,'Thursday':4,'Friday':5,'Saturday':6,'Sunday':7})
train_temp['Traffic Situation'] = train_temp['Traffic Situation'].replace({'low':1,'normal':2,'high':3, 'heavy':4})
train_temp['Time'] = le.fit_transform(train_temp['Time'])
                                                

train = train_temp

train
```

```python
test_temp = test
test_temp['Day of the week'] = test_temp['Day of the week'].replace(
    {'Monday': 1, 'Tuesday': 2, 'Wednesday': 3, 'Thursday': 4, 'Friday': 5, 'Saturday': 6, 'Sunday': 7})
test_temp['Traffic Situation'] = test_temp['Traffic Situation'].replace(
    {'low': 1, 'normal': 2, 'high': 3, 'heavy': 4})

test_temp['Time'] = le.fit_transform(test_temp['Time']) 

test = test_temp

test
```

```python
X_test = test.drop(columns=['Traffic Situation'],axis=1)
y_test = test['Traffic Situation']

X_test

X_train
```

```python
X_train = train.drop(columns=['Traffic Situation'], axis=1)
y_train = train['Traffic Situation']

model = LogisticRegression()

lr = model.fit(X_train,y_train)

```

```python
y_predict = model.predict(X_test)

accuracy = accuracy_score(y_test,y_predict)
print("Accuracy of the model : ", accuracy*100)
```
